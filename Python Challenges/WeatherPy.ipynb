```python
from citipy import citipy
import json
import requests
import pandas as pd
from api_keys import weather_api_key
from api_keys import g_key
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.stats import linregress
```

## Generate Cities List

```python
# Range of latitudes and longitudes
latitude_range = (-90.000, 78.2166667)
longitude_range = (-179.3, 179.5)
```

```python
# create an empty lists for both cities and latitude and longitude
cities =[]
lat_lng = []

Latitude = np.random.uniform(low=-54.8,high=78.2166667,size = 1200)
Longitude = np.random.uniform(low=-179.3,high=179.5,size = 1200)
lat_lng=zip(Latitude,Longitude)
```

```python
#tuple(lat_lng)
```

```python
for latlng in lat_lng:
    city = citipy.nearest_city(latlng[0],latlng[1]).city_name
    
    if city not in cities:
        cities.append(city)

print(len(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = 'http://api.openweathermap.org/data/2.5/weather?'

#http://api.openweathermap.org/data/2.5/weather?appid=25bc90a1196e6f153eece0bc0b0fc9eb&units=imperial&q=chokurdakh
```

```python
city_data = []
print('Beginning Data Retrieval')
print('-----------------------------------------')
set_count = 1
record_count=1
for i, city in enumerate(cities):
    # to create batch of records and each batch holds 50 records
    if(i % 50 == 0 and i >=50):
        set_count += 1
        record_count = 0
    
    print('Processing Record {0} of Set {1} | {2}'.format(record_count,set_count,city))
    
    full_url = url + 'appid=' + weather_api_key + '&units=imperial&q=' +city
    #print(full_url)

    record_count+=1
    
    try:
        response = requests.get(full_url).json()
        # collect the city data
        city_latitute = response["coord"]["lat"]
        city_longitude = response["coord"]["lon"]
        city_temperature = response['main']['temp_max']
        city_humidity = response['main']['humidity']
        city_clouds = response['clouds']['all']
        city_wind = response['wind']['speed']
        city_country = response['sys']['country']
        city_dt = response['dt']

        # append collected city data to city_data list
        city_data.append({"City": city, 
                          "Lat": city_latitute, 
                          "Lng": city_longitude, 
                          "Max Temp": city_temperature,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_dt
                         })
    except:
        print('city not found')
        pass
   
print('-----------------------------')
print('Data Retrieval Complete')
print('-----------------------------')
```

### Convert Raw Data to DataFrame

## Linear Regression

```python
North_Hem_df = city_data_df[city_data_df['Lat'] >= 0]
North_Hem_df
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = North_Hem_df['Lat']
y_values = North_Hem_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
print('The r-rvalue is: %s' % rvalue)
plt.show()
```
